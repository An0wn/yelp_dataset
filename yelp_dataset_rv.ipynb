```python
df = pd.read_csv('yelp_reviews_AZ_Phoenix.csv')
print(df.shape)
df.sample(3)
```

```python
# Check if review text mentioned 'RV ' or ' RV' that is case sensitive
df_text_RV = df[df['text'].str.contains('RV | RV', case=True, na=False)].name.value_counts()
df_text_RV[:10]
```

```python
# https://twitter.com/justmarkham/status/1161621998994374657
# Create one row for each series that contain comma-separated items
df_explode = df.assign(categories = df.categories.str.split(', ')).explode('categories')
df_explode.sample(3)
```

```python
print('Total number of categories:', len(df_explode.categories.value_counts()))
df_explode[df_explode['categories'].str.contains('RV', na=False)].categories.value_counts()
```

```python
# Check if business categories are RV related
df_RV_cat = df[df['categories'].str.contains('RV Repair|RV Dealers|RV Rental|RV Parks|Campgrounds', case=False, na=False)]
print(df_RV_cat.shape)
df_RV_cat.head()
```

```python
df_RV_explode = df_RV_cat.assign(categories = df_RV_cat.categories.str.split(', ')).explode('categories')
df_RV_explode[df_RV_explode['categories'].str.contains('RV', na=False)].categories.value_counts()
```
